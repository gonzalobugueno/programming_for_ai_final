```python
plt.figure(figsize=(10, 10))
plt.ticklabel_format(axis='both', style='plain')
plt.ylim(0, 800_000)
plt.yticks(range(0, 800001, 20000))
plt.ylabel("Amount of the given credit")
plt.xlabel("Age")
for defaulted, dd in df.groupby('default payment next month'):
    plt.scatter(dd['AGE'], dd['LIMIT_BAL'], label="Yes" if defaulted else "No", alpha=0.5, s=10)
plt.legend(title="Defaulted next week")
plt.show()
```

There's a correlation between lower credit and defaulting, independently of age

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

dd = df[df['default payment next month'] == True]

heatmap_data, x_edges, y_edges = np.histogram2d(
    dd['AGE'], dd['LIMIT_BAL'], bins=[range(df['AGE'].min(), df['AGE'].max() + 1, 1), range(0, 530_001, 10_000)]
)

plt.figure(figsize=(10, 10))

sns.heatmap(
    heatmap_data.T,
    xticklabels=x_edges,
    yticklabels=y_edges,
    cbar_kws={'label': 'Count'}
)

plt.title("Heatmap of age vs credit (Defaulted next week)")
plt.xlabel("Age of client")
plt.ylabel("Credit, TWD")
plt.show()
```

There is more frequency of credit at TWD 20.000, 50.000, 200.000, 360.000, and 500.000, suggesting the bank may be employing "steps" or "levels"

## Training model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[["LIMIT_BAL", "SEX", "EDUCATION", "MARRIAGE", "AGE", "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6",
        "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6", "PAY_AMT1", "PAY_AMT2",
        "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]].copy()

y = df["default payment next month"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

rf = RandomForestClassifier(random_state=42, n_estimators=100, min_samples_split=10, min_samples_leaf=1,
                            max_features='sqrt', max_depth=30)
rf.fit(X_train, y_train)

rf.score(X_test, y_test)
```

## Load into DB

```python
import sqlite3

con = sqlite3.connect('../project.db')
cur = con.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS defaultd(
    limit_bal INT NOT NULL,
    sex INT NOT NULL,
    education INT NOT NULL,
    marriage INT NOT NULL,
    age INT NOT NULL,
    pay_0 INT NOT NULL,
    pay_2 INT NOT NULL,
    pay_3 INT NOT NULL,
    pay_4 INT NOT NULL,
    pay_5 INT NOT NULL,
    pay_6 INT NOT NULL,
    bill_amt1 INT NOT NULL,
    bill_amt2 INT NOT NULL,
    bill_amt3 INT NOT NULL,
    bill_amt4 INT NOT NULL,
    bill_amt5 INT NOT NULL,
    bill_amt6 INT NOT NULL,
    pay_amt1 INT NOT NULL,
    pay_amt2 INT NOT NULL,
    pay_amt3 INT NOT NULL,
    pay_amt4 INT NOT NULL,
    pay_amt5 INT NOT NULL,
    pay_amt6 INT NOT NULL,
    label INT NOT NULL
);
""")

for (idx, row), label in zip(X.iterrows(), y):
    cur.execute("""
    INSERT INTO defaultd (limit_bal,sex,education,marriage,age,pay_0,pay_2,pay_3,pay_4,pay_5,pay_6,bill_amt1,bill_amt2,bill_amt3,bill_amt4,bill_amt5,bill_amt6,pay_amt1,pay_amt2,pay_amt3,pay_amt4,pay_amt5,pay_amt6,label)
    VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
    """, [
        row['LIMIT_BAL'],
        row['SEX'],
        row['EDUCATION'],
        row['MARRIAGE'],
        row['AGE'],
        row['PAY_0'],
        row['PAY_2'],
        row['PAY_3'],
        row['PAY_4'],
        row['PAY_5'],
        row['PAY_6'],
        row['BILL_AMT1'],
        row['BILL_AMT2'],
        row['BILL_AMT3'],
        row['BILL_AMT4'],
        row['BILL_AMT5'],
        row['BILL_AMT6'],
        row['PAY_AMT1'],
        row['PAY_AMT2'],
        row['PAY_AMT3'],
        row['PAY_AMT4'],
        row['PAY_AMT5'],
        row['PAY_AMT6'],
        label
    ])

con.commit()
con.close()
```
